```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer as SIA
import pandas as pd
pd.set_option('display.max_colwidth', -1)
```

# Bullish tweets

```python
data = pd.read_csv("../Dumps Twitter/bullish_tweets.csv")
columns_to_drop = ["Created At","Verified User","Reply Tweet",
                   "Reply To User ID","Original Tweet","Tweet ID",
                   "Favorite Count"
                  ]
data.drop(columns_to_drop, axis=1, inplace=True)
print("{} bullish tweets".format(data.shape[0]))
```

```python
analyzer = SIA()
data["Sentiment"] = data.apply(lambda row: analyzer.polarity_scores(row["Status"])["compound"], axis=1)
errors = data.loc[data["Sentiment"] < 0]
nb_errors = errors.shape[0]
print("Accuracy : {:2.2%} ({} errors)".format(1 - nb_errors/data.shape[0], nb_errors))
```

### Errors : bullish tweets with a negative sentiment

```python
display(errors)
```

# Bearish tweets

```python
data = pd.read_csv("../Dumps Twitter/bearish_tweets.csv")
columns_to_drop = ["Verified User","Created At","Tweet ID",
                   "Favorite Count","Original Tweet",
                   "Reply Tweet","Reply To User ID"
                  ]
data.drop(columns_to_drop, axis=1, inplace=True)
print("{} bearish tweets".format(data.shape[0]))
```

```python
analyzer = SIA()
data["Sentiment"] = data.apply(lambda row: analyzer.polarity_scores(row["Status"])["compound"], axis=1)
errors = data.loc[data["Sentiment"] > 0]
nb_errors = errors.shape[0]
print("Accuracy : {:2.2%} ({} errors)".format(1 - nb_errors/data.shape[0], nb_errors))
```

### Errors : bearish tweets with a positive sentiment

```python
display(errors)
```
