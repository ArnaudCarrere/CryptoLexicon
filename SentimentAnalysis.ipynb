```python
from vaderSentiment.vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer as SIA
import pandas as pd
pd.set_option('display.max_colwidth', -1)
```

```python
data = pd.read_csv("Dumps Twitter/bullish_tweets.csv")
columns_to_drop = ["Created At","Verified User","Reply Tweet",
                   "Reply To User ID","Original Tweet","Tweet ID",
                   "Favorite Count"
                  ]
data.drop(columns_to_drop, axis=1, inplace=True)
display(data)
```

```python
analyzer = SIA()
data["Sentiment"] = data.apply(lambda row: analyzer.polarity_scores(row["Status"])['compound'], axis=1)
display(data)
```
